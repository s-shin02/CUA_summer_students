```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [5,6,7,8]

print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.title('x vs y', fontsize = 20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array(x)
y = np.array(y)
m,b = np.polyfit(x,y,1)

plt.scatter(x,y)
plt.plot(x, m*x + b)
plt.title('x vs y', fontsize = 20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
excluded_index = [1]
x_new = [i for n, i in enumerate(x) if n not in excluded_index]

excluded_index = [2]
y_new = [i for n, i in enumerate(y) if n not in excluded_index]

print(x_new)
print(y_new)
```

```python
x_new = np.array(x_new)
y_new = np.array(y_new)

x = np.array(x)
y = np.array(y)
m,b = np.polyfit(x,y,1)

plt.scatter(x,y)
plt.scatter(x_new,y_new)
plt.plot(x, m*x + b)
plt.title('x vs y', fontsize = 20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
m,b = np.polyfit(x,y,1)
m1,b1 = np.polyfit(x_new, y_new, 1)

plt.scatter(x,y)
plt.scatter(x_new,y_new)
plt.plot(x, m*x + b)
plt.plot(x_new, m1*x_new + b1)
plt.title('x vs y', fontsize = 20)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
